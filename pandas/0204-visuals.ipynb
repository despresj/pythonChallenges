```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyprojroot import here

df = pd.read_csv(here("pandas/data/my_gm.csv"))
```

```python
df.columns
```

```python
plt.scatter(df['gdpPercap'], df['lifeExp'])
```

```python
plt.scatter(np.log(df['gdpPercap']), df['lifeExp'])
plt.title("The relationship appears to be much more log-linear")
plt.xlabel('logGdp')
```

```python
plt.scatter(np.log10(df['gdpPercap']), df['lifeExp'])
plt.title("The relationship appears to be much more log-linear")
plt.xlabel('log10Gdp')
```

```python
m, b = np.polyfit(np.log10(df['gdpPercap']), df['lifeExp'], 1)
y = b + m*np.log10(df['gdpPercap'])
plt.scatter(np.log10(df['gdpPercap']), df['lifeExp'],
            alpha = 0.2, s = 3)
plt.plot(np.log10(df['gdpPercap']), y, color="r")

plt.title("The relationship appears to be much more log-linear")
plt.xlabel('log10Gdp')
```
